# **Instrues funcionamento da API**

### **1 - Conexo com o Oracle DB**

Na funo *conexao_db()* voc precisa passar suas credenciais do banco de dados Oracle. Para maior segurana:

1- crie um arquivo *credenciais.json* na na mesma pasta do projeto *server.py*

2- Adicione suas credenciais

Com isso voc ir conectar ao seu banco de dados Oracle.




### **2 - Criando as tabelas no banco**

 necessrio que voc tenha a tabela criada no seu banco com produtos dentro dela.

Para isso, voc pode usar o arquivo *criar_tabelas.sql* que est na pasta *scripts* do projeto.



__1-__ Copie o Create Table e cole no Sql developer



__2-__ Clique no boto play, no canto superior esquerdo do script



__3-__ Logo aps voc precisar clicar no boto para commitar as alteraes



__4-__ Voc pode conferir se os produtos foram adicionados, executando a query *SELECT*, contida na pasta scripts





### **3 - Ligando o servidor e retornando a o contedo da API**

Agora com o banco de dados criado e com dados nele, podemos fazer as requies no banco e gerar um arquivo JSON, via API, para ser consumido no endpoint

__1-__ Abrir o terminal e dar o comando __*"cd .\server\\"*__ e depois o comando __*pip install requirements.txt*__

__2-__ Entrar no arquivo *server.py*__

__3-__ Abra o terminal e escreva ***python server.py***
    
    
Isso inicializar o servidor Flask no navegador

__4-__ Acessar o navegador e produrar por ***http://localhost:8080/api/produtos***


__5-__ Ao acessar o navegador, dever ser retornado um arquivo JSON, com todos os produtos cadastrados no seu banco de dados
